```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_blobs
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Instantiate dataset with 4 random clusters
X, y_true = make_blobs(n_samples=300, centers=4,cluster_std=0.60, random_state=0)


# TODO determine the best k for k-means
# TODO calculate accuracy for best K

graph = KMeans()
visual = KElbowVisualizer(graph,k = (1,15))
visual.fit(X)
visual.show()


newGraph = KMeans(4)
newGraph.fit(X)
yPrediction = newGraph.predict(X)
print("The accuracy score is: ", accuracy_score(y_true, yPrediction))

# TODO draw a confusion matrix
sns.set()
matrix = confusion_matrix(y_true,yPrediction)
sns.heatmap(matrix, square = True, annot = True, fmt ='d',cbar = False)
plt.xlabel('True Label')
plt.ylabel('Predicted Label')
```
