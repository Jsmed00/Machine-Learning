```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest


# load the breast_cancer dataset
init_data = load_breast_cancer()
(X, y) = load_breast_cancer(return_X_y=True,as_frame=True)

print(X.shape)

```

```python
X.sample(5)
```

```python
# split data into input and taget variable(s)

x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# TODO Select some features (X), hint: based on the connections with
# our Y (importance? correlation?)

n_features = 2
k_best = SelectKBest(k=n_features)
x_train_new = k_best.fit_transform(x_train, y_train)
x_test_new = k_best.transform(x_test)
```

```python
# TODO Tune parameters for RandomForestClassifier

rand_classifier = RandomForestClassifier(class_weight='balanced', random_state=0)

#define parameters for gridsearch

param_grid = {
    'n_estimators': [20,50,100,200],
    'min_samples_split': [2,4],
    'min_samples_leaf': [1,2,4],
    'criterion': ["gini", "entropy"],
}
```

```python
# TODO need 5 fold cross validation
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(rand_classifier, param_grid, verbose=2)
grid.fit(x_train_new, y_train)

print("Best parameters: ", grid.best_params_)

model = grid.best_estimator_
y_pred = model.predict(x_test_new)
```

```python
# TODO Calculate Average accuracy score

acc = accuracy_score(y_test, y_pred)

# TODO Calculate Average (accuracy score/number of features)
acc_per_feature = acc / n_features

print("Accuracy: {}, per feature: {}".format(acc, acc_per_feature))
```
